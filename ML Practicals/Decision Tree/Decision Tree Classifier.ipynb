```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
iris=load_iris()
```

```python
iris
```

```python
iris.data
```

```python
iris.target
```

```python
X_train,X_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=0.3)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
# Calaculate accuracy of the model
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True)
```

```python
# Grid Search CV
from sklearn.model_selection import GridSearchCV
```

```python
param_grid= {
    'max_depth':[3, 4, 5, 6],
    'min_samples_split':[2, 3, 4, 5],
    'min_samples_leaf':[1, 2, 3]
}
```

```python
grid_search=GridSearchCV(estimator=model,param_grid=param_grid,cv=5)
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_params_
```

```python
model2=DecisionTreeClassifier(max_depth=3,min_samples_leaf=1,min_samples_split=4)
```

```python
model2.fit(X_train,y_train)
```

```python
y_predict=model2.predict(X_test)
```

```python
accuracy_score(y_test,y_predict) #here accuracy is redcuced becuause previously the accuracy we got was when we bulit the tree to its complete depth
# But in this case we reduced overfitting of the data by pruning it . training accuracy got changed
```
