## <span style="color :Red;">DBSCAN Clustering</span>

```python
# Importing the necessary Libraries
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#DB scan is mainly used to cluster non_linear data
X,y=make_moons(n_samples=250,noise=0.05) #The make_moons dataset is a swirl pattern, or two moons.two half circles interleaving with each other.
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
#Feature Scaling(standard scaling)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
## DBSCAN Algorithm
```

```python
dbscan=DBSCAN(eps=0.5)
```

```python
dbscan.fit(X_scaled)
```

```python
dbscan.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```
